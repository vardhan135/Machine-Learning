```python
import numpy as np

def mandelbrot(c,maxiter):
    z=c
    for n in range(maxiter):
        if abs(z) > 2:
            return n
        z = z*z + c
        return 0
def mandelbrot_set(xmin,xmax,ymin,ymax,width,height,maxiter):
    r1 = np.linespace(xmin, xmax, width)
    r2 = np.linespace(ymin, ymax, height)
    r3 = np.empty((width,height))
    for i in range(width):
            for j in range(height):
                r3[i,j] = mandelbrot(r1[i] + 1j*r2[j],maxiter)
                abc = r1
                return(r1,r2,r3)   
```

```python
#Ensembles for improving the accuracy using Baggingclassifier
import pandas as pd
from pandas import read_csv
import numpy as np
import sklearn
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
splits = 10
seed = 7
cart = DecisionTreeClassifier()
model = BaggingClassifier(base_estimator=cart,n_estimators=100,random_state=seed)
kfold = KFold(n_splits=splits,random_state=seed)
results = cross_val_score(model,X,Y,cv=kfold)
print('Accuracy',results.mean()*100)
```

```python
#Ensembles for improving the accuracy using RandomForestClassifier
import pandas as pd
from pandas import read_csv
import sklearn
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
seed=7
splits=10
model = RandomForestClassifier(n_estimators=100,max_features=3,random_state=seed)
kfold = KFold(n_splits=splits,random_state=seed)
results = cross_val_score(model,X,Y,cv=kfold)
print('Accuracy:',results.mean()*100)
```

```python
#ensembles for improving accuracy using boosting like Adaboost and gradientboosting
import pandas
from pandas import read_csv
import sklearn
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
seed=7
splits=10
model_ada = AdaBoostClassifier(n_estimators=splits,random_state=seed)
model_gbc = GradientBoostingClassifier(learning_rate=0.1,max_features=3,random_state=seed)
kfold = KFold(n_splits=splits,random_state=seed)
result_ada = cross_val_score(model_ada,X,Y,cv=kfold)
result_gbc = cross_val_score(model_gbc,X,Y,cv=kfold)
print('Accuracy:Ada',result_ada.mean()*100,'Accuracy:Gradient:',result_gbc.mean()*100)
```

```python
#ensembles for improving accuracy using voting
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
#Creating Pipeline with FeatureSelection&Union along with model
import pandas as pd
from pandas import read_csv
import sklearn
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
seed=7
splits=10
models = []
models.append(('LogisticRegression',LogisticRegression()))
models.append(('SVM',SVC()))
models.append(('DecisionTree',DecisionTreeClassifier()))
model = VotingClassifier(models)
kfold = KFold(n_splits=splits,random_state=seed)
result = cross_val_score(model,X,Y,cv=kfold)
print('Voting Accuracy:',result.mean()*100)
```

```python
#Improving performance with algorithm tuning using Grid Search
import numpy as np
import pandas as pd
import sklearn
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
alphas = np.array([1,0.1,0.01,0.001,0])
param_grid = dict(alpha=alphas)
model = Ridge()
grid = GridSearchCV(estimator=model,param_grid=param_grid)
grid.fit(X,Y)
print(grid.best_score_)
print(grid.best_estimator_.alpha)
```

```python
#Improving performance using algorithm tuning using Randomized search
import pandas as pd
import numpy as np
import sklearn
import scipy
from sklearn.linear_model import Ridge
from scipy.stats import uniform
from sklearn.model_selection import RandomizedSearchCV
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
param_grid = {'alpha':uniform()}
model = Ridge()
rsearch = RandomizedSearchCV(estimator=model,param_distributions=param_grid,n_iter=100,random_state=7)
rsearch.fit(X,Y)
print(rsearch.best_score_)
print(rsearch.best_estimator_.alpha)
```

```python
#Saving ML algorithm to a file
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from pickle import dump
from pickle import load
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.7,test_size=0.3)
model = LogisticRegression()
model.fit(X_train,Y_train)
filename = 'finalized_model.sav'
dump(model,open(filename,'wb'))
```

```python
loaded_model = load(open('finalized_model.sav','rb'))
result = loaded_model.score(X_test,Y_test)
print(result.mean())
```

```python
#Saving ML algorithm to a file using job
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.externals.joblib import dump
from sklearn.externals.joblib import load
df = pd.read_csv("C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv")
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.7,test_size=0.3,random_state=7)
model = LogisticRegression()
model.fit(X_train,Y_train)
filename = 'finalized_model2.sav'
dump(model,filename)
```
