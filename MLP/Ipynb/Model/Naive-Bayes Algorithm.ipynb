```python
# Naive-Bayes: Used mostly for text classification and for data having multiple classes.

#   Naive-Bayes Regressor for regression, and Naive-Bayes Classifier for classification problems

# This algorithm considers that features are not related
# It can be used for both regresssion and classification problems - for predicting the labeled data using bayes probability
# Used for large data sets for text classification and mulitple classes for a label.

# pros -
# 1. It is fast and easy to predict class of test data.
# 2. If the data features have no relationship with one another then it is better than most sophisticated algorithms

# cons-
# 1. Most features in real data are related 
# 2. For regression problems it assumes normal distribution
# 3. It is considered as bad estimator with less significant probabilities

# Applications:

# 1. Real-Time Prediction: It is used for real time prediction, it is fast and accurate on categorical data.
# 2. Multi-Class : If the labeled data has multiple classes to be predicted Naive-bayes is fast at prediction.
# 3. Can be used extensively for text-classification, sentiment analysis and spam filtering.
# 4, can be used for recommendation prediction.


# Models:
# Three types of Naive-Bayes in SciKit
# 1. Gaussian - assumes the data is normal
# 2. Multinomial - Bernoulli (data is 0 and 1's)

# Improve model:
# 1. If continuous features doesnot have normal distribution we need to convert them into normal
# 2. Remove zero frequency issue by applying laplace correction
# 3. Remove corelated features.
# 4. Ensemble, bagging, boosting won't help improve the accuracy
```

```python
# Nive-Bayes Classifier for predicting diabetes (Feature Selection and Preprocessing)

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import sklearn
from sklearn.naive_bayes import GaussianNB
df = pd.read_csv('C:\\Users\\vardh\\Vardhan\\ED\\diabetes.csv')
df.head()
df.dtypes
df.describe()
df.corr()
df.dropna()
df.groupby('Outcome').count()
df.head()
```

```python
# Normalizing the data - Since all the values are continuous 
from sklearn.preprocessing import Normalizer
pp =Normalizer()
td = pp.fit_transform(df)
td[0,:]
td = pd.DataFrame(td)
```

```python
# Feature extraction
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
names = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DPF','Age']
model = SelectKBest(score_func=chi2,k=4)
fit = model.fit(X,Y)
scores = pd.DataFrame(fit.scores_,names)
print(scores)
X.head()
```

```python
X = df.iloc[:,[1,4,5,7]]
Y = df.iloc[:,8]

from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
N = Normalizer()
fit = N.fit_transform(X,Y)
#print(fit)
td = pd.DataFrame(fit)
#print(td)
X = td
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=7)
NB = GaussianNB()
fit1 = NB.fit(X_train,Y_train)
score = cross_val_score(NB,X_train,Y_train)
print(score.mean())
new_data = [['150','100','25.5','36'],['85','120','15','95'],['100','95','35','45']]
pred_data = N.fit_transform(new_data)
prediction = fit1.predict(pred_data)
for i in prediction:
    if prediction[i] == 0:
        print('No Diabetes')
    else:
        print('Diabetes')
```
