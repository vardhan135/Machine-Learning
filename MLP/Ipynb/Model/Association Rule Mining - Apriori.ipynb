```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
data = pd.read_excel('C:\\Users\\vardh\\Vardhan\\ED\\OnlineRetail.xlsx')
data.head()
```

```python
data.dropna()
data_france = (data[data.Country == 'France'].groupby(['InvoiceNo','Description',])['Quantity']
               .sum().unstack().reset_index().fillna(0)
               .set_index('InvoiceNo'))
data_france
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
data_france = data_france.applymap(encode_units)
data_france.drop('POSTAGE',inplace=True,axis=1)

```

```python
frequent_items = apriori(data_france,min_support = 0.07,use_colnames=True)
association = association_rules(frequent_items,metric='lift',min_threshold=1)
association.head()
```
